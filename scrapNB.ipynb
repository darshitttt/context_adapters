```python
import os
import json
import pandas as pd
from pydub import AudioSegment
from datetime import datetime
from tqdm import tqdm

def convert_to_seconds(time_val):
    if isinstance(time_val, pd.Timestamp):  # Convert if it's a pandas Timestamp
        time_val = time_val.strftime("%H:%M:%S.%f")
    
    dt = datetime.strptime(time_val, "%H:%M:%S.%f")
    return dt.hour * 3600 + dt.minute * 60 + dt.second + dt.microsecond / 1_000_000


# Function to segment audio
def segment_audio(row, output_dir):
    audio_path = row['audio_file']
    transcription_path = row['transcription_file']
    
    # Load audio
    audio = AudioSegment.from_file(audio_path)
    
    # Load transcription JSON
    with open(transcription_path, 'r') as f:
        transcription_data = json.load(f)
    
    segments = []
    start_time = None
    end_time = None
    text = ""
    segment_counter = 0
    trimmed_audio_files = []
    
    for entry in transcription_data:
        #print(entry['start_time'])
        cur_start = convert_to_seconds(entry['start_time'])
        cur_end = convert_to_seconds(entry['end_time'])
        cur_text = entry['words']

        if start_time is None:
            start_time = cur_start
            end_time = cur_end
            text = cur_text
        else:
            duration = cur_end - start_time

            # If the duration is within the limit, keep adding words
            if duration < 30:
                end_time = cur_end
                text += " " + cur_text
            else:
               
                
                #print((end_time-start_time), start_time, end_time)
                # Trim audio
                trimmed_audio = audio[start_time * 1000:end_time * 1000]
                trimmed_audio_path = os.path.join(output_dir, f"trimmed_{row['audio_file'].split('/')[-1].split('.')[0]}_{(int)(start_time)}_{(int)(end_time)}.wav")
                trimmed_audio.export(trimmed_audio_path, format="wav")
                
                segments.append([trimmed_audio_path, text, (end_time-start_time)])
                trimmed_audio_files.append(trimmed_audio_path)
                
                # Reset variables for the next segment
                start_time = cur_start
                end_time = cur_end
                text = cur_text
                segment_counter += 1

    # Handle the last segment
    if start_time and text:
        trimmed_audio = audio[start_time * 1000:]
        trimmed_audio_path = os.path.join(output_dir, f"trimmed_{row['audio_file'].split('/')[-1].split('.')[0]}_{(int)(start_time)}_{(int)(end_time)}.wav")
        trimmed_audio.export(trimmed_audio_path, format="wav")
        segments.append([trimmed_audio_path, text, (end_time-start_time)])
        trimmed_audio_files.append(trimmed_audio_path)

    return segments

# Main processing
def process_dataframe(df, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    all_segments = []

    for _, row in tqdm(df.iterrows(), desc='Processing audios ...', total=len(df)):
        segments = segment_audio(row, output_dir)
        all_segments.extend(segments)

    # Create new DataFrame
    new_df = pd.DataFrame(all_segments, columns=['trimmed_audio', 'transcription', 'duration'])
    new_df.to_csv(os.path.join(output_dir, "chime6_segmented_audio.csv"), index=False)

    return new_df

# Example Usage
'''df = pd.DataFrame({
    'audio_file': ['audio1.wav', 'audio2.wav'],
    'transcription_file': ['transcription1.json', 'transcription2.json']
})'''
#dff = df[0:1]

output_directory = "/ceph/dpandya/chime6/chime6_eval/"
new_df = process_dataframe(df, output_directory)
#print(new_df.head())
```
